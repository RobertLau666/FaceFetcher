```python
from pypinyin import pinyin, Style

def chinese_to_pinyin(name):
    # 调用 pinyin 函数将中文名字转换为拼音，默认风格为 'TONE2'
    pinyin_list = pinyin(name, style=Style.NORMAL)
    # 将拼音列表连接成字符串，并将首字母大写
    pinyin_str = ''.join([item[0] for item in pinyin_list]).capitalize()
    return pinyin_str

# 测试
chinese_name = "张三"
pinyin_name = chinese_to_pinyin(chinese_name)
print("中文名字：", chinese_name)
print("拼音：", pinyin_name)
```
